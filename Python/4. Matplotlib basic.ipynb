# Matplotlib

- pyplot 객체를 사용하여 데이터를 표시
- pyplot 객체에 그래프들을 쌓은 다음 plt.show()를 이용하여 출력

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline  

X = np.arange(100)
Y = X**2
plt.plot(X, Y)
plt.show()
```

- Graph는 기본 Figure 객체에 생성됨 (따로 설정하지 않으면 기본 Figure에 그래프 누적됨)
- pyplot 객체 사용 시, 기본 Figure에 그래프가 그려짐

```python
x1 = np.arange(100)
y1 = np.sin(x1)

x2 = np.arange(100)
y2 = np.cos(x2)

plt.plot(x1, y1)
plt.plot(x2, y2)
plt.show()
```

## Figure 지정

- Figure 지정하여 그래프 그리기
- Figure ⊃ Axes ⊃ Graph

```python
fig = plt.figure()
fig.set_size_inches(10, 5) # 크기 지정
ax1 = fig.add_subplot(1, 2, 1) # 행 갯수, 열 갯수, 몇 번째
ax2 = fig.add_subplot(1, 2, 2)

ax1.plot(x1, y1, c = "b")
ax2.plot(x2, y2, c = "g")
plt.show()
```

## LineStyle 지정

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline  

x = np.arange(100)
y1 = 0.5 * x
y2 = 2 * x
plt.plot(x, y1, lineStyle = "dashed")
plt.plot(x, y2, lineStyle = "dotted")
```

## Title 지정

- Pyplot에 title 함수 사용, figure의 subplot 별 입력 가능

```python
plt.plot(x, y1, color = "b", linestyle = "dashed")
plt.plot(x, y2, color = "r", linestyle = "dotted")
plt.title("Two lines")
plt.xlabel("X value")
plt.ylabel("Y value")
plt.show()
```

- Latex 타입의 표현도 가능 (수식 표현 가능)

```python
plt.title("$y = \\frac{a}{b}x + c$")
plt.show()
```

## legend(도표 범례) 작성하기

- legend 함수로 범례를 표시함, loc 위치 등 속성 지정

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline  

x = np.arange(100)
y1 = 0.5 * x
y2 = 2 * x

plt.plot(x, y1, label = "$y = \\frac{1}{2}x$")
plt.plot(x, y2, label = "$y = 2x$")
plt.legend(shadow = True, fancybox = True, loc = "upper right")
```

## 기타 그래프 꾸미기

- Graph에 텍스트 및 기호 넣기

```python
plt.plot(x, y1, label = "$y = \\frac{1}{2}x$")
plt.plot(x, y2, label = "$y = 2x$")
plt.legend(shadow = True, fancybox = True, loc = "upper right")
plt.text(80, 30, "Line1")
plt.annotate("Line2", xy=(60, 125), xytext=(40, 130), 
             arrowprops = dict(facecolor = "black", shrink = 0.05))
plt.show()
```

- Graph에 격자 넣기
- xy축 범위 조정하기

```python
plt.plot(x, y1, label = "$y = \\frac{1}{2}x$")
plt.plot(x, y2, label = "$y = 2x$")
plt.legend(shadow = True, fancybox = True, loc = "upper right")
plt.text(80, 30, "Line1")
plt.annotate("Line2", xy=(60, 125), xytext=(40, 130), 
             arrowprops = dict(facecolor = "black", shrink = 0.05))

# Graph에 격자 넣기
plt.grid(linewidth = 0.5, linestyle = "--")
# xy축 범위 조정하기
plt.xlim(-50, 120)
plt.ylim(-50, 250)

plt.show()
```
