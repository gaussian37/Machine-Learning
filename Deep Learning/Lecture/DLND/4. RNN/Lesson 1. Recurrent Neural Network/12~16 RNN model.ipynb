RNN은 Recurrent Neural Network의 줄임말 입니다. 여기서 Recurrent라는 용어를 사용하는 이유는 무엇일까요? RNN은 각각의 Input 에 대하여 동일한 작업을 계속 Neural Network에 적용하기 때문입니다.

![2](http://blogfiles.naver.net/MjAxODAxMjNfMjA4/MDAxNTE2NzE4MjIwNDE2.zFbGJMFDc49tSAXS0rRsS2PwJ7EK53sri4ualF3V7bUg.NHIqJ4nFdV5m5uXohXEEkZM-it3qLnumyaAutZ5h0Ggg.PNG.infoefficien/12._RNN%28a%29.mp4_000054648.png?type=w1)

RNN은 반복 작업을 위하여 이전 input을 저장하기 위해 내부에 memory를 따로 두게 되는데 이것을 state라고 합니다. 

![3](http://blogfiles.naver.net/MjAxODAxMjNfMTk4/MDAxNTE2NzE4Mzg0MDU1.RVWoA1MU-FzpwUzTGhnkU6JzszIlgGYK8CkXC2mWfIYg.EijPuALCvyCkDNrVccSED7rLAFWgHXjREHt2Y2rzYZ4g.PNG.infoefficien/12._RNN%28a%29.mp4_000078194.png?type=w1)

많은 Application들이 Temporal Dependency(시간 의존성) 성질을 가집니다. 즉, 현재 output($\overline{y}$)은 현재 input($\overline{x}$) 뿐만 아니라 과거 input이 고려된 memory($\overline{s}$) 또한 사용되어 집니다. 

![4](http://blogfiles.naver.net/MjAxODAxMjNfMTMx/MDAxNTE2NzE4NTg4NzY5.NAulX9PlaNu4ljR69J1UWAFvzgubVOjhG_-2QNyHa_Yg.IofOYzinWLLKmnzAcLIwQo9-uu3FbArXflXj0kJYzLEg.PNG.infoefficien/12._RNN%28a%29.mp4_000086993.png?type=w1)

현재 Input과 과거 Input을 이용하는 RNN 구조의 대표적인 예는 문장에서 다음 단어를 예측하는 것입니다. To be or not to _ 에서 다음 단어를 예측할 때 현재 Input은 to 이고 과거 Input은 to 이전 단어들 입니다.  

RNN 또한 기본적으로 FFNN(Feedforward Neural Network)의 구조를 가집니다. 
RNN과 기존의 Feedforward NN의 차이점에 대하여 살펴보겠습니다.

![5](http://blogfiles.naver.net/MjAxODAxMjNfMjk5/MDAxNTE2NzE5MjAxNDI4.kbn2X-9r6N3GrSOsY60JUTPbxmEo9ecsSA-XAGG4sEUg.gLC8sWpE2W8vSm0rEZe3ZAz2TaNhywjXqeLPGvD4mLYg.PNG.infoefficien/12._RNN%28a%29.mp4_000170857.png?type=w1)

첫 번째로, <span class="mark">input과 output을 정의하는 방식에 차이</span>가 있습니다.<br> FFNN(Feedforward Neural network)에서는 각각의 step에서 단일 input, 단일 output을 사용 하였습니다. 반면 RNN에서는 input과 output은 many-to-many, many-to-one, one-to-many 구조 등으로 가져갈 수 있습니다. <br>
두번째로,<span class="mark">현재 input 뿐만 아니라 memory에 해당하는 과거 input의 연산 결과가 다음 step의 input</span>으로 주어집니다. 

![6](http://blogfiles.naver.net/MjAxODAxMjNfMjE5/MDAxNTE2NzE5NDkwNjQ4.b-nmft9V474oyleq8o8XOW5fYNIsqFqKdnVqSdntUmUg.3fLnKej8egrbpjeW_Otp66kY-zZAVM9FDWwSFYop7G4g.PNG.infoefficien/12._RNN%28a%29.mp4_000190029.png?type=w1)

FFNN에서는 Input → Output 경로에서 어떤 feedback도 찾을 수 없었습니다. 

![7](http://blogfiles.naver.net/MjAxODAxMjRfMjM3/MDAxNTE2NzE5NjUxNDI1.j3Y-6sIizwnVQVB3MRrG397vY467DjtCuyvXM-_fTJsg.lJ3SY2fS9BCAQodjEFep0rpk4KI8doW8xgGk8I3sYDAg.PNG.infoefficien/12._RNN%28a%29.mp4_000219041.png?type=w1)

앞에서 정의하였던 state의 구조를 보면 <span class="mark">State Machine</span> 처럼 볼 수 있습니다. 사실 state라는 용어도 state machine 구조와 유사하여 가져온 것입니다. <span class="mark">state machine에서는 현재의 input과 이전 step의 output이 동시에 input이 되어 새로운 output이 도출되는 구조입니다.</span>RNN에서는 이 구조 뿐만 아니라 Network 스스로 weight를 어떻게 최적화 할지에 대한 train을 수행하게 됩니다.
